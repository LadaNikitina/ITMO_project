# Импорты

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')
```

# Чтение данных

```python
df = pd.read_csv('downstream_data.csv', index_col = 'Unnamed: 0')
display(df.head())
```

```python
df.info()
```

```python
df.describe(include='all')
```

# Аналитика

```python
print("Число пропусков по столбцам:")
print(df.isnull().sum())
```

```python
print("Уникальные метки:", df['label'].unique())
print("Уникальные сплиты:", df['split'].unique())
print("Уникальные языки:", df['language'].unique())
print("Уникальные названия датасетов:", df['dataset_name'].unique())
```

```python
plt.figure(figsize=(6,4))
df['split'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Распределение по сплитам')
plt.ylabel('Количество примеров')
plt.xlabel('Сплит')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['language'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Распределение по языкам', fontsize=16)
plt.ylabel('Количество примеров', fontsize=14)
plt.xlabel('Язык', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,4))
df['dataset_name'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Распределение по датасетам')
plt.ylabel('Количество примеров')
plt.xlabel('Название датасета')
plt.show()
```

```python
# Аналитика по длине реплик
df['length_chars'] = df['utterance'].apply(len)
df['length_words'] = df['utterance'].apply(lambda x: len(str(x).split()))

df[['length_chars', 'length_words']].describe()
```

```python
plt.figure(figsize=(8,4))
df['length_chars'].hist(bins=30)
plt.title('Распределение по числу символов')
plt.xlabel('Число слов')
plt.ylabel('Частота')
plt.show()
```

```python
plt.figure(figsize=(8,4))
df['length_words'].hist(bins=30)
plt.title('Распределение по числу слов')
plt.xlabel('Число слов')
plt.ylabel('Частота')
plt.show()
```

```python
for dataset_name in df['dataset_name'].unique():
    subset = df[df['dataset_name'] == dataset_name]
    
    print()
    print(f"Аналитика для датасета: {dataset_name}")
    print("Количество примеров:", len(subset))
    print("Распределение по меткам:")
    print(subset['label'].value_counts())
    
    plt.figure(figsize = (8, 4))
    subset['length_chars'].hist(bins = 20)
    plt.title(f'Распределение длины реплик (в символах) - {dataset_name}')
    plt.xlabel('Число слов')
    plt.ylabel('Частота')
    plt.show()
    
    plt.figure(figsize = (8, 4))
    subset['length_words'].hist(bins = 20)
    plt.title(f'Распределение длины реплик (в словах) - {dataset_name}')
    plt.xlabel('Число слов')
    plt.ylabel('Частота')
    plt.show()
```

```python
# Анализ частотности слов для всего датасета
stop_words = set(
    stopwords.words('english') + stopwords.words('russian') + \
    stopwords.words('spanish') + stopwords.words('french') + \
    stopwords.words('german') + stopwords.words('italian')
)

def get_tokens(text):
    tokens = word_tokenize(str(text).lower())
    tokens = [token for token in tokens if token.isalpha() and token not in stop_words and len(token) > 2]
    return tokens

df['tokens'] = df['utterance'].apply(get_tokens)

all_tokens = [token for tokens in df['tokens'] for token in tokens]
word_freq = Counter(all_tokens)

print("Топ-20 самых частотных слов во всем датасете:")
print(word_freq.most_common(20))
```

```python
# Облако слов для всего датасета
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(all_tokens))
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Облако слов - Все датасеты")
plt.show()
```

```python
plt.figure(figsize = (12, 6))
sns.boxplot(data = df, x = 'language', y = 'length_words')
plt.title('Длина реплик (в словах) по языкам')
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (12, 6))
sns.boxplot(data = df, x = 'language', y = 'length_chars')
plt.title('Длина реплик (в символах) по языкам')
plt.xticks(rotation = 45)
plt.show()
```
